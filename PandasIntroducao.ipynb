```python
# Importando a biblioteca pandas
import pandas as pd
```

```python
# Criando um dataframe a partir de um dicionário
#dataframe = pd.DataFrame()
venda ={
    'data': ['10/02/1998','16/03/2012'],
    'valor': [250,890],
    'produto': ['Macarrão', 'Arroz'],
    'quantidade': [150,250]
}

print(venda)

vendas_df = pd.DataFrame(venda)
print(vendas_df)

# Display é o mais recomendado para visualização
display(vendas_df)
```

#### Importando um arquivo Excel

```python
tabela_excel = pd.read_excel('Vendas.xlsx')
display(tabela_excel)
```

#### Métodos úteis 

- heade - shape - describe

```python
display(tabela_excel.head(10)) # Visualizando as 10 primeiras linhas da tabela
```

```python
print(tabela_excel.shape) # Visualizando a quantidade de linhas e colunas 
```

```python
# Informação geral, de média, contagem, desvio padrão e muito mais
display(tabela_excel.describe())
```

#### Editando nosso Data Frame

```python
# pegandp só a coluna produto, uma única coluna é uma série do pandas
produtos = tabela_excel[['Produto', 'ID Loja']]
display(produtos)
```

Pegando linhas especificas, métodos .loc, que també é uma série do pandas.

```python
# pegar uma linha
print("Pegando uma linha")
display(tabela_excel.loc[9]) # Pegando uma linha
print("\n \n")
print("Pegando duas linhas")
display(tabela_excel.loc[[1,5]]) # Pegando duas linhas
```

```python
# pegando da linha 10 até a linha 20
display(tabela_excel.loc[10:20])
```

Pegando todas as linhas que correspondem a uma condição

```python
# E vamos armazenar em uma variável Produto_Relogio
Produto_Relogio = display(tabela_excel.loc[tabela_excel['Produto'] == 'Relógio'])
```
